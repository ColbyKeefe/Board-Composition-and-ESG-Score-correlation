```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
finance_df = pd.read_excel('ESQ Board Composition Data.xlsx', header = None)
finance_df = finance_df.reset_index(drop=True)
finance_df.head()
```

```python
Finance_Table = pd.concat([JP_Morgan, Morgan_Stanley, BOA, GS, City])
Finance_Table.head()
```

```python
X = Finance_Table[['Board Size', 'Independent Members', 'Board Independence', '# of Females']]
y_governance = Finance_Table['Governance Score']
y_esg_overall = Finance_Table['ESG Score Overall']
y_combined_esg = Finance_Table['Combined ESG Score']
X = X.astype('float')
y_governance = y_governance.astype('float')
y_esg_overall = y_esg_overall.astype('float')
y_combined_esg = y_combined_esg.astype('float')
model_esg_governance = sm.OLS(y_governance, sm.add_constant(X)).fit()
model_esg_overall = sm.OLS(y_esg_overall, sm.add_constant(X)).fit()
model_combined_esg = sm.OLS(y_combined_esg, sm.add_constant(X)).fit()
print(model_esg_governance.summary())
print(model_esg_overall.summary())
print(model_combined_esg.summary())
```

```python
print('Governance P-Value:', model_esg_governance.pvalues)
print('Overall ESG PV-Value:', model_esg_overall.pvalues)
print('Combined ESG P-Value:', model_combined_esg.pvalues)
```
